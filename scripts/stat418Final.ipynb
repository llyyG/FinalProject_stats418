```
## Avengers: Endgame

# Make the request
AV4_raw = requests.get('https://www.the-numbers.com/movie/Avengers-Endgame-(2019)#tab=box-office').text

# Turn into soup, specify the HTML parser
AV4_soup = BeautifulSoup(AV4_raw,'html.parser')

# Use .find_all to retrieve all the tables in the page
AV4_tables = AV4_soup.find_all('table')

AV4_trs = AV4_tables[12].find_all('tr')

cleaned_AV4_rows = []

# Loop through all the non-header (first row) table rows
for row in AV4_trs[1:]:
    
    # Get the text of the row and split on the newlines (like above)
    cleaned_row = row.text.split('\n')
    
    # Add this cleaned row back to the external list of row data
    cleaned_AV4_rows.append(cleaned_row)
    
# Inspect the first few rows of data
cleaned_AV4_rows[:2]

AV4_df = pd.DataFrame(cleaned_AV4_rows)

# Inspect
# AV4_df.head()

# Drop columns 0 and 11 and overwrite the xmas_box_df variable
AV4_df = AV4_df.drop(columns=[0,9])

# Rename the columns
AV4_df.columns = ['Date','Rank','Gross','% Change','# Theaters',
                      'Per Theater','Total gross','Days']

# Write to disk
# AV4_df.to_csv('AV4_box_office.csv',encoding='utf8')

# Inspect
AV4_df.head()
```

```
from sklearn.metrics import mean_squared_error
from pyramid.arima import auto_arima
train_diff = gross_AV4 - gross_AV4.shift()
train_diff = train_diff.dropna().values
model_tarima = auto_arima(train_diff, trace=True, error_action='ignore', suppress_warnings=True)
model_tarima.fit(train_diff)
```

```
from statsmodels.tsa.arima_model import ARIMA
model_train = ARIMA(train_diff, order=(1, 0, 0))  
results_train = model_train.fit() 

plt.plot(train_diff)
plt.plot(results_train.fittedvalues, color='red')
```
